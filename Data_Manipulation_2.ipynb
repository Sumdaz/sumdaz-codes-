__7) Describe the outliers by box plot: Write a a loop for the object columns of df3 whose number of unique values is less than 10.  <font color='purple'> 15 points </font>__

```python
for column in df3.select_dtypes(include='object'):
    if df3[column].nunique() < 10:
        plt.figure(figsize=(12,8))
        ax=sns.boxplot(y=column, x='Price', data=df3)
        plt.show()
```

__8) Describe the correlation of df3: <font color='purple'> 10 points </font>__

```python
df3.corr()
desc = df3.corr(method = 'pearson')

#the Upper triagular corr-matrix
trimask = np.triu(np.ones_like(desc,dtype =bool))
fig = plt.figure(figsize=(10,6))
# using the upper triangular matrix as mask
sns.heatmap(desc, cmap='RdYlGn_r', mask=trimask|(np.abs(desc)>=0.5),annot=True)
```

__11) Is there any signaficant relation between Regionname and Price? - <font color='purple'> 20 points </font>__

```python
from scipy.stats import f_oneway
#Create the Data Frame
price_reg = df[['Regionname', 'Price']]
price_reg
```

```python
# Finds out the Prices data for each Regionname as a list
compare=price_reg.groupby('Regionname')['Price'].apply(list)

result = f_oneway(*compare)
result
```
