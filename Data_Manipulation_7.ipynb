#### Q1) Using groupby(), plot the number of songs that have been released each year. Please, use the kind='bar', xlabel=year and ylabel ='Count of songs' and the size of figure should be (12,8) ---- <font color='blue'> (15 points)</font>

```python
song_yearly = songs.groupby('year').size().reset_index(name='count')
c = ['red', 'yellow', 'black', 'blue', 'orange', 'green', 'purple']
plt.figure(figsize=(12,8))
plt.bar(song_yearly['year'], song_yearly['count'], color = c)
plt.xlabel('Year')
plt.ylabel('Count of songs')
plt.title('Number of yearly released songs')
plt.show()
```

#### Q2) Use groupby() to present how many songs does Rihanna have over years. <font color='blue'> (10 points)</font>

```python
#to get the songs by Rihanna in each year
songs_by_Rihanna = songs[songs['artist'] == 'Rihanna'].groupby('year').count()

songs_by_Rihanna = songs_by_Rihanna.reset_index()

# filter out the year and song columns
songs_by_Rihanna = songs_by_Rihanna [['year', 'song']]
songs_by_Rihanna
```

#### Q3) Use groupby() to sort the number of artists released pop(only!) songs in 2011. <font color='blue'> (10 points)</font>

```python
#to select rows where genre is pop and year is 2011
pop_songs_in_2011 = songs[(songs['genre']=='pop') & (songs['year']== 2011)]
```

#### Q6) <font color='blue'> TOTAL: (50 points)</font>
* Define a new datafame which contains the popularity, danceability, energy, year, tempo, song.  <font color='blue'> (5 points)</font>
* Use transform() to define a new column which represent percantage of popularity (pop%). <font color='blue'> (10 points)</font>
* Set the index of dataframe as year assign new data frame as song_rearranged<font color='blue'> (10 points)</font>
* Describe a new column as 'score' by considering mean of popularity, danceability, energy constants. <font color='blue'> (10 points)</font>
* Then, use groupby() in years and use apply function which returns the sorted values with respect to 'score' column? <font color='blue'> (15 points)</font>
* Display the songs having the first 4 best scores in 2009. Please be sure there is no duplication in the data! <font color='blue'> (10 points)</font>

```python
#Define a new datafame which contains the popularity, danceability, energy, year, tempo, song.
my_songs = songs[['popularity', 'danceability', 'energy', 'year', 'tempo', 'song']]
my_songs.sample(10)
```

```python
#Use transform() to define a new column which represent percantage of popularity (pop%).
my_songs['pop%'] = my_songs['popularity'].transform(lambda x: x/x.sum()*100)
my_songs.sample(10)
```

```python
# Set the index of dataframe as year assign new data frame as song_rearranged
song_rearranged = my_songs.set_index('year')
song_rearranged.sample(10)
```

```python
#Describe a new column as 'score' by considering mean of popularity, danceability, energy constants.
song_rearranged['score'] = (song_rearranged['popularity'] + song_rearranged['danceability'] + song_rearranged['energy'])/3
song_rearranged.sample(10)
```

```python
#Then, use groupby() in years and use apply function which returns the sorted values with respect to 'score' column? (15 points)
sorted_scores =song_rearranged.groupby('year').apply(lambda x: x.sort_values(by="score", ascending=False))
sorted_scores.sample(10)
```

```python
# Display the songs having the first 4 best scores in 2009. Please be sure there is no duplication in the data! 
song_rearranged=song_rearranged.reset_index()
best_top_4_songs_in_2009 = song_rearranged[song_rearranged['year'] == 2009].sort_values('score', ascending=False).drop_duplicates().head(4)
best_top_4_songs_in_2009[['song', 'score']]
```
